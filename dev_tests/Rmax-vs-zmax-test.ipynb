```python
# Import DYNAMITE
import dynamite as dyn
```

```python
# Generate a few models or connect to existing ones
fname = 'user_test_config_ml.yaml'
c = dyn.config_reader.Configuration(fname,
                                        reset_logging=True,
                                        user_logfile='test_nnls',
                                        reset_existing_output=False)
_ = dyn.model_iterator.ModelIterator(c)
```

```python
# Verify that there are models
c.all_models.table.pprint
```

```python
# Instantiate Plotter object
p = dyn.plotter.Plotter(c)
```

```python
# Feel free to experiment...

p.rmax_zmax_plot(model=None,
                 max_r=None,
                 max_z=None,
                 col_scale='log',
                 components='all',
                 figtype='.png')

p.rmax_zmax_plot(col_scale='log', components='each')

p.rmax_zmax_plot(col_scale='log', components=['thin_d', 'bulge']);
```
